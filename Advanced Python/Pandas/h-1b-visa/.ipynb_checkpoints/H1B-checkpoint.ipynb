```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('../data/h-1b-visa/h1b_kaggle.csv')
```

```python
col  = list(df.columns)
print(col)
```

```python
df.drop(['Unnamed: 0','lon','lat'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
col = list(df.columns)
col
```

```python
df.rename({'CASE_STATUS':'status',
          'EMPLOYER_NAME':'company',
           'JOB_TITLE':'job',
          'FULL_TIME_POSITION':'time',
          'PREVAILING_WAGE':'salary'}
         , inplace=True,axis=1)
```

```python
df.columns
```

```python
df.rename({"WORKSITE":'state'},inplace=True,axis=1)
```

```python
df.columns
```

```python
year_df = df.groupby('YEAR')['YEAR'].count()
```

```python
plt.title("Number of Application by Year",fontsize=20)
year_df.plot(kind='bar')
plt.show()
```

```python
df['status'].unique()
```

```python
df[df['status'].isin(['INVALIDATED',
    'PENDING QUALITY AND COMPLIANCE REVIEW - UNASSIGNED'])].shape
```

```python
df = df[~(df['status'].isin(['INVALIDATED',
    'PENDING QUALITY AND COMPLIANCE REVIEW - UNASSIGNED']))]
```

```python
df['status'].unique()
```

```python
v = df['status'].value_counts()
```

```python
plt.title("CASE STATUS")
v.plot(kind='bar')
plt.show()
```

```python
x = df.groupby('YEAR')['status'].value_counts()
```

```python
type(x)
```

```python
x.shape
```

```python
year = [ 2011,2012,2013,2014,2015,2016]
for each_year in year : 
    x[each_year].plot(kind='bar',title=f"Year {each_year}")
    plt.legend()
    plt.show()
```
