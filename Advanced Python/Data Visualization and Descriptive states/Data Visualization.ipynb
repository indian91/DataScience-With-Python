```python
x = np.arange(1,11)
y = np.random.randint(20,50,10)
plt.scatter(x,y)
plt.plot([x.mean(),x.mean()],[y.min(),y.max()],'r',lw=3)
plt.plot([x.mean()-x.std()]*2,[y.min(),y.max()],'g',lw=3)
plt.plot([x.mean()+x.std()]*2,[y.min(),y.max()],'g',lw=3)
for index in range(len(x)):
    plt.plot([x[index],x.mean()],[y[index]]*2,'m--',lw=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
from matplotlib.animation import FuncAnimation
%matplotlib notebook 
x = list(np.arange(1,11))
y = list(np.random.randint(10,25,10))
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y,'r^:')

import time
def change(interval):
    global y
    y.pop(0)
    y.append(np.random.randint(10,25))
    time.sleep(interval)
    ax.clear()
    ax.plot(x,y,'r^:')
    
ani = FuncAnimation(fig,change,2)
plt.show()
```

```python
import psutil
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
from matplotlib.animation import FuncAnimation
import time
fig=plt.figure()
ax=fig.add_subplot(111)
y=[psutil.cpu_percent() for var in range(50)]
x=[var for var in range(50)]
ax.plot(x,y)

def change(interval):
    global x,y
    x.pop(0)
    y.pop(0)
    x.append(x[-1]+1)
    y.append(psutil.cpu_percent())
    ax.clear()
    ax.plot(x,y,'r^--')
ani=FuncAnimation(fig,change,2000)
plt.show()
```
