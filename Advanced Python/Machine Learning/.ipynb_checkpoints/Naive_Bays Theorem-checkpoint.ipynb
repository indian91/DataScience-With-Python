### Evaluation on IRIS datasets

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
type(iris)
```

```python
dir(iris)
```

```python
print(iris['DESCR'])
```

```python
iris['data'][:5]
```

```python
iris['feature_names']
```

```python
iris['target'][:5]
```

```python
iris['target_names']
```

```python
features=pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
features[:5]
```

```python
target=pd.Series(iris['target']).map({0:'setosa',1:'versicolor',2:'virginica'})
```

```python
target[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.30)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[:5]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### Evaluation on DogvsCat datasets

```python
import cv2
```

```python
from PIL import Image
```

```python
import os
```

```python
cats_image=[]
path='dogvscat/training_set/cats'
for image in os.listdir(path):
    full_path=os.path.join(path,image)
    img=cv2.imread(full_path,cv2.IMREAD_GRAYSCALE)
    img=cv2.resize(img,(128,128))
    cats_image.append(img)
    cats=np.concatenate(cats_image)
```

```python
dogs_image=[]
path='dogvscat/training_set/dogs'
for image in os.listdir(path):
    full_path=os.path.join(path,image)
    img=cv2.imread(full_path,cv2.IMREAD_GRAYSCALE)
    img=cv2.resize(img,(128,128))
    dogs_image.append(img)
    dogs=np.concatenate(dogs_image)
```

```python
df1=pd.DataFrame(cats)
df2=pd.DataFrame(dogs)
```

```python
df1['label']='cat'
df2['label']='dog'
```

```python
df=pd.concat([df1[:5000],df2[507000:]])
```

```python
df.head()
```

```python
df.tail()
```

```python
features=df.drop('label',axis=1)
```

```python
target=df['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target)
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
df  = pd.read_csv('navie_bays_play.csv')
```
