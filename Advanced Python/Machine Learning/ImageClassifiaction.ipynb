```python
clean_images=[]
path='images/train/clean'
for image in os.listdir(path):
    full_path=os.path.join(path,image)
    img=cv2.imread(full_path,cv2.IMREAD_GRAYSCALE)
    img=cv2.resize(img,(128,128))
    clean_images.append(img)
```

```python
clean=np.concatenate(clean_images)
```

```python
clean.shape
```

```python
df1=pd.DataFrame(clean)
```

```python
messy_images=[]
path='images/train/messy'
for image in os.listdir(path):
    full_path=os.path.join(path,image)
    img=cv2.imread(full_path,cv2.IMREAD_GRAYSCALE)
    img=cv2.resize(img,(128,128))
    messy_images.append(img)
```

```python
messy=np.concatenate(messy_images)
```

```python
messy.shape
```

```python
df2=pd.DataFrame(messy)
```

```python
df1['label']=0
```

```python
df2['label']=1
```

```python
df=([df1,df2])
```

```python
df1.shape
```

```python
df2.shape
```

```python
df1.head()
```

```python
df2.head()
```

```python
df=pd.concat([df1,df2])
```

```python
df.head()
```

```python
df.tail()
```

```python
feature=df.drop('label',axis=1)
target=df['label']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
model=LogisticRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,target)
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
y_pred=model.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
